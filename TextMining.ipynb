```python
import pandas as pd
import numpy as np
import re
from collections import Counter
from konlpy.tag import Okt
okt = Okt()
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pickle
import time
from sklearn.feature_extraction.text import CountVectorizer # 빈도 기준
```

```python
raw = pd.read_csv('C:/Users/naval/Desktop/sh/result_sanghyun.csv')
```

```python
def strip_e(st):
    try:
        RE_EMOJI = re.compile('[\U00010000-\U0010ffff]', flags=re.UNICODE)
        return RE_EMOJI.sub(r'', st)
    except:
        return ''
```

```python
raw
```

```python
#이게 한 동꺼 frequency
dong_freq_list = []
for dong in raw['0'].unique():
    temp = raw[raw['0']==dong]
    comment_col = temp.columns[7:]
    dong_comments = []
    for col in comment_col:
        tmp = temp[col].tolist()
        for i in range(len(tmp)):
            tmp[i] = strip_e(tmp[i])
            tmp[i] = okt.morphs(tmp[i],norm=True,stem=True)
            tmp[i] = ' '.join(tmp[i])
            dong_comments.append(tmp[i])
    try:
        count_vect = CountVectorizer()
        cvect = count_vect.fit_transform(dong_comments)
        dtme = pd.DataFrame(cvect.todense(), columns=count_vect.get_feature_names())
        freq = dtme.sum(axis=0).sort_values()
        dong_freq_list.append([dong,freq])
    except:
        pass
```

```python
len(dong_freq_list)
```

```python
with open('dong_freq_list.pickle', 'wb') as handle:
    pickle.dump(dong_freq_list, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#이게 한 동꺼 frequency
stopwords = ['하다', '시키다', '좋다', '너무', '보다', '자다', '오다', '있다', '같다',
            '이다', '진짜', '주문', 'ㅎㅎ', '으로', '잇다', '않다', '없다',
             '항상', '감사하다', '피자', '넘다', '족발', '정말', '괜찮다', '잇다',
            '보쌈', '음식', '인데', '갈비', '물회', '하고', '주문', ]
dong_freq_list = []
for dong in raw['0'].unique():
    temp = raw[raw['0']==dong]
    comment_col = temp.columns[7:]
    dong_comments = []
    for col in comment_col:
        tmp = temp[col].tolist()
        for i in range(len(tmp)):
            tmp[i] = strip_e(tmp[i])
            tmp[i] = okt.morphs(tmp[i],norm=True,stem=True)
            tmp[i] = ' '.join(tmp[i])
            dong_comments.append(tmp[i])
    try:
        tfidf = TfidfVectorizer(max_features=2000, stop_words=stopwords, max_df=0.4)
        tivect = tfidf.fit_transform(dong_comments).todense()
        dtme = pd.DataFrame(tivect, columns=tivect.get_feature_names())
        freq = dtme.sum(axis=0).sort_values()
        dong_freq_list.append([dong,freq])
    except:
        pass
```
